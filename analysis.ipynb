Since they are many missing values, we need to find the relationship between missing values and Sales Price
Let's plot some diagram for this relationship

```python
for feature in features_with_na:
    data = df.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
for feature in features_with_na:
    data = df.copy()
    
    # Create a new column indicating if the value is missing (1) or not (0)
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # Plot the count of missing vs non-missing for each category in SalePrice
    plt.figure(figsize=(8, 5))
    sns.countplot(data=data, x='RainTomorrow', hue=feature)
    plt.title(f"Relationship between missing values in {feature} and RainTomorrow")
    plt.show()
```

```python
# Step 1: Identify features with more than 30% missing values
missing_percentage = df.isnull().mean() * 100
features_with_na_30 = missing_percentage[missing_percentage > 30].index

# Step 2: Handle missing values by filling them (e.g., with -1 or 0 for categorical, or mean for numeric)
data = df.copy()
for feature in features_with_na_30:
    # If the feature is categorical, you can fill it with -1 or other placeholder
    if data[feature].dtype == 'object':  # For categorical features
        data[feature] = data[feature].fillna('Missing')  # Use 'Missing' as a placeholder
    else:  # For numeric features
        data[feature] = data[feature].fillna(data[feature].mean())  # Fill with mean or median

# Step 3: Encode 'RainTomorrow' if necessary (assuming 'RainTomorrow' is categorical)
data['RainTomorrow'] = data['RainTomorrow'].map({'Yes': 1, 'No': 0})

# Step 4: Calculate the correlation with 'RainTomorrow' for these features
correlation_matrix = data[features_with_na_30].corrwith(data['RainTomorrow']).sort_values(ascending=False)

# Step 5: Display the correlation values
print("Correlation with RainTomorrow for features with more than 30% missing values:")
print(correlation_matrix)

# Optional: Visualize the correlation values
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.barplot(x=correlation_matrix.index, y=correlation_matrix.values)
plt.xticks(rotation=90)
plt.title("Correlation of Features with RainTomorrow (More than 30% Missing Values)")
plt.ylabel("Correlation Coefficient")
plt.xlabel("Features")
plt.show()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=df[feature].median()
    
    ## create a new feature to capture nan values
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_with_nan].isnull().sum()
```

```python
df=df.dropna()  #just for now
df.shape
```

```python
#Encoding the data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df=df.apply(le.fit_transform)
```

```python
# # Dividing dependent and independent variables
X = df.drop("RainTomorrow", axis=1)
y = df["RainTomorrow"]
```

```python
'''
# Sampling
from imblearn.over_sampling import SMOTE 
from collections import Counter

sampler = SMOTE()
X, y = sampler.fit_resample(X, y)

print('SMOTE {}'.format(Counter(y)))
'''
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression  

classifier= LogisticRegression(random_state=0)  
lr_model=classifier.fit(X_train, y_train)
y_pred=lr_model.predict(X_test)
```

```python
# Accuracy = TP+TN/TP+FP+FN+TN
# Precision = TP/TP+FP
# Recall = TP/TP+FN
#F1 Score = 2*(Recall * Precision) / (Recall + Precision)


from sklearn.metrics import classification_report, confusion_matrix 
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['No','Yes']
y_axis_labels = ['No','Yes']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

clf= KNeighborsClassifier(n_neighbors=25)

# Train the model using the training sets
knn_model=clf.fit(X_train,y_train)

y_pred=knn_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['No','Yes']
y_axis_labels = ['No','Yes']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='viridis', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```
