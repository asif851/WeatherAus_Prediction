```python
df= df[df['RainTomorrow'].notna()]
df.head()
```

```python
df['RainTomorrow'].value_counts()
```

```python
# finding categorical variables in the df
categorical = df.select_dtypes(include=['object']).columns.tolist()
print('categorical columns:', categorical)
```

```python
df[categorical].head()
```

```python
# List of columns to fill missing values
categorical_columns = ['WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday']

# Loop through each column and fill missing values with the mode
for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# finding numerical variable 
numerical_features = df.select_dtypes(include=['number']).columns.tolist()
print('Number of numerical variables: ', len(numerical_features))

print('Numerical columns are:',numerical_features)
```

```python
# Numerical variables are usually of 2 type
# Continous variable and Discrete Variables

discrete_feature=[feature for feature in numerical_features if len(df[feature].unique())<20]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
print('discrete_feature columns are:',discrete_feature)
```

```python
for feature in ['Cloud9am', 'Cloud3pm']:
    df[feature].fillna(df[feature].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Identify continuous features
continuous_features = [feature for feature in numerical_features if feature not in discrete_feature]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap for missing values in continuous features
sns.heatmap(df[continuous_features].isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values in Continuous Features")
plt.show()
```

```python
for feature in continuous_features:
    df[feature].fillna(df[feature].median(), inplace=True)

print("Missing values in continuous variables handled using median imputation.")
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#Encoding the data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df=df.apply(le.fit_transform)
df.head()
```

```python
# # Dividing dependent and independent variables
X = df.drop("RainTomorrow", axis=1)
y = df["RainTomorrow"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression  

classifier= LogisticRegression(random_state=0)  
lr_model=classifier.fit(X_train, y_train)
y_pred=lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix 
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['No','Yes']
y_axis_labels = ['No','Yes']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

clf= KNeighborsClassifier(n_neighbors=25)

# Train the model using the training sets
knn_model=clf.fit(X_train,y_train)

y_pred=knn_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['No','Yes']
y_axis_labels = ['No','Yes']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True, xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```
