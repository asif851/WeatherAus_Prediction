Since they are many missing values, we need to find the relationship between missing values and RainTomorrow

```python
# Step 1: Identify features with more than 30% missing values
missing_percentage = df.isnull().mean() * 100
features_with_na_30 = missing_percentage[missing_percentage > 30].index

# Step 2: Handle missing values by filling them (e.g., with -1 or 0 for categorical, or mean for numeric)
data = df.copy()
for feature in features_with_na_30:
    # If the feature is categorical, you can fill it with -1 or other placeholder
    if data[feature].dtype == 'object':  # For categorical features
        data[feature] = data[feature].fillna('Missing')  # Use 'Missing' as a placeholder
    else:  # For numeric features
        data[feature] = data[feature].fillna(data[feature].mean())  # Fill with mean or median

# Step 3: Encode 'RainTomorrow' if necessary (assuming 'RainTomorrow' is categorical)
data['RainTomorrow'] = data['RainTomorrow'].map({'Yes': 1, 'No': 0})

# Step 4: Calculate the correlation with 'RainTomorrow' for these features
correlation_matrix = data[features_with_na_30].corrwith(data['RainTomorrow']).sort_values(ascending=False)

# Step 5: Display the correlation values
print("Correlation with RainTomorrow for features with more than 30% missing values:")
print(correlation_matrix)

# Optional: Visualize the correlation values
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.barplot(x=correlation_matrix.index, y=correlation_matrix.values)
plt.xticks(rotation=90)
plt.title("Correlation of Features with RainTomorrow (More than 30% Missing Values)")
plt.ylabel("Correlation Coefficient")
plt.xlabel("Features")
plt.show()
```

Feature Selection
information gain method

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif

# Assuming the data has already been loaded as DataFrame `X` (features) and `y` (target)
# Replace `X` and `y` with your actual feature DataFrame and target variable.

# Calculate mutual information
importances = mutual_info_classif(X, y)

# Create a dictionary with feature names and their importance values
feature_importance_dict = {feature: importance for feature, importance in zip(X.columns, importances)}

# Print all feature importances
print("Feature Importances:")
for feature, importance in feature_importance_dict.items():
    print(f"'{feature}': {importance}")

# Filter features with importance >= 0.01
filtered_features = {feature: importance for feature, importance in feature_importance_dict.items() if importance >= 0.01}

# Print filtered features
print("\nFeatures with Importance >= 0.01:")
for feature, importance in filtered_features.items():
    print(f"'{feature}': {importance}")

# Create a bar plot for feature importances (not sorted)
plt.figure(figsize=(10, 8))
feat_importances = pd.Series(importances, index=X.columns)
feat_importances.plot(kind="barh", color="skyblue")
plt.title("Feature Importances")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

```python

# Sampling
from imblearn.over_sampling import SMOTE 
from collections import Counter

sampler = SMOTE()
X, y = sampler.fit_resample(X, y)

print('SMOTE {}'.format(Counter(y)))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression  

classifier= LogisticRegression(random_state=0)  
lr_model=classifier.fit(X_train, y_train)
y_pred=lr_model.predict(X_test)
```

```python
# Accuracy = TP+TN/TP+FP+FN+TN
# Precision = TP/TP+FP
# Recall = TP/TP+FN
#F1 Score = 2*(Recall * Precision) / (Recall + Precision)


from sklearn.metrics import classification_report, confusion_matrix 
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['No','Yes']
y_axis_labels = ['No','Yes']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

clf= KNeighborsClassifier(n_neighbors=25)

# Train the model using the training sets
knn_model=clf.fit(X_train,y_train)

y_pred=knn_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['No','Yes']
y_axis_labels = ['No','Yes']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='viridis', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize an empty list to store accuracy scores
accuracy_scores = []

# Loop over a range of k values
for k in range(1, 51):  # Testing k values from 1 to 50
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train, y_train)  # Train the model
    
    # Predict on the test set
    y_pred = clf.predict(X_test)
    
    # Calculate the accuracy and append it to the list
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append((k, accuracy))

# Find the k value with the highest accuracy
best_k, best_accuracy = max(accuracy_scores, key=lambda x: x[1])

# Print the results
print(f"Optimal k: {best_k} with accuracy: {best_accuracy}")

# Optional: Visualize accuracy scores
import matplotlib.pyplot as plt

k_values = [score[0] for score in accuracy_scores]
accuracies = [score[1] for score in accuracy_scores]

plt.plot(k_values, accuracies, marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Number of Neighbors (k)')
plt.show()
```
